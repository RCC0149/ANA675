```python
import numpy as np
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Part a: Design a Backpropagation neural network to store the following letters: (30 points)

# Training data
X = np.array([[1,1,1,0,1,0,0,0,0],  # T
              [1,0,1,1,1,1,1,0,1],  # H
              [1,1,1,1,0,0,1,1,1]]) # C
y = np.array([[1,0,0],  # T
              [0,1,0],  # H
              [0,0,1]])  # C

# Build the model
model = Sequential([
    Dense(6, activation='relu', input_shape=(9,)),  # Hidden layer with 6 neurons
    Dense(3, activation='softmax')                  # Output layer with 3 classes
])

# Compile the model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
model.fit(X, y, epochs=100, verbose=0)

# Show model summary of the designated NN (5 points)
model.summary()
```

```python
# Part b: Input the following shape and let me know if the neural network can recognize this letter or not. (25 points)

new_shape = np.array([[1,0,1, 1,0,1, 0,1,0]])
prediction = model.predict(new_shape)
print(prediction)
```

No, the neural network cannot recognize this letter.
